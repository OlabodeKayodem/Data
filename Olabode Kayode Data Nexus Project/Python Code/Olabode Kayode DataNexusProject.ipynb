# Step 1: Data Cleaning

```python
#Import Neccessary Python Libaries for data manipulation, analysis, and computation.

import pandas as pd # -"panel data," an econometrics term for multidimensional structured data sets.
import numpy as np # - "Numerical Python" a fundamental package for scientific computing in Python.
```

```python
# Read the CSV file into a DataFrame called 'df', Note:df can be any variable.
df = pd.read_csv('C:\\Users\\user\\Desktop\\Python\\bank_marketing.csv')
```

```python
# Display the DataFrame bytyping the variable 'df'
df
```

```python
# Display the first 50 rows of the DataFrame, this allows you to inspect the data and its structure
df.head(50)
```

```python
# Display the last 50 rows of the DataFrame, to inspect the tail end of the dataset. Note:By default it shows last 5!
df.tail(50)
```

```python
# Add a new column 'year' with value 2022 to the DataFrame.
df['year'] = 2022
```

```python
# Generate 'campaign_Id' based on index and append/add it as a new column.
df['campaign_Id'] = ['Id{}'.format(i+1) for i in range(len(df))]
```

```python
# Display the DataFrame to confirm that the added columns 'year & campaign_Id' appear.
df
```

```python
# Map the month names to numeric values and create a 'Date' column.
# Reason: To standardize date representations in a DataFrame, making it easier to work with date-related operations and analysis.
month_mapping = {
    'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 'jun': 6,
    'jul': 7, 'aug': 8, 'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12
}
df['month'] = df['month'].str.lower().map(month_mapping)
df['Date'] = pd.to_datetime(df[['year', 'month', 'day']])
df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')

# Hey, just a heads up: it's usually a good idea to stick with the Year-Month-Day date format, since that's what most databases prefer. Keep it sleek and consistent!
```

```python
# Display the DataFrame to confirm that the added columns 'Date' appear.
df
```

```python
# Display information about the DataFrame, including data types and missing values.
df.info()
```

```python
#Since we now have the Year in the right format, we need to drop the columns for Month, Day, and Year, reducing our total number of columns to 22.
columns_to_drop = ['month', 'day', 'year'] 
df = df.drop(columns=columns_to_drop)
```

```python
# Display the DataFrame to confirm that the Month, Day, and Year, columns have been dropped.
df
```

```python
#Display the count of missing values in each column of the DataFrame
df.isnull().sum()
```

# Step 2: Data Preprocessing

```python
# We need to split the data set with a schema comprising three tables: "Client", "Campaign", and "Economics".
```

```python
# Create a DataFrame containing selected columns for the 'Client' table.
client_Table = df[['client_id', 'age', 'job', 'marital', 'education', 'credit_default','housing','loan']]
```

```python
# Display the 'Client' DataFrame
client_Table
```

```python
# Display information about the 'Client' DataFrame, including data types and missing values
client_Table.info()
```

```python
# Make a copy of the DataFrame to avoid chained assignment.
# Note: Chained assignment, also known as chained indexing, occurs when you perform multiple indexing or attribute access operations in a single statement. This can lead to unexpected behavior and should generally be avoided, especially when setting values in a DataFrame.
client_Table_copy = client_Table.copy()
```

```python
# Define mapping dictionary and handle the unknown variable with the numpy libarary
mapping = {'yes': True, 'no': False, 'unknown': np.nan}

# Convert columns using map and astype on the copied DataFrame
client_Table_copy['credit_default'] = client_Table_copy['credit_default'].map(mapping).astype(bool)
client_Table_copy['housing'] = client_Table_copy['housing'].map(mapping).astype(bool)
client_Table_copy['loan'] = client_Table_copy['loan'].map(mapping).astype(bool)
```

```python
#Display information about the modified 'Client' DataFrame to confirm that 'credit_default', 'housing', and 'loan' have been converted to BOOLEAN as requested.
client_Table_copy.info()
```

```python
# Make a copy of the DataFrame to avoid chained assignment.
client_Table_copy1 = client_Table_copy.copy()
```

```python
# Rename the 'client_id' column to 'id'
client_Table_copy1.rename(columns={'client_id': 'id'}, inplace=True)
```

```python
# Display the modified 'Client' DataFrame with renamed column.
client_Table_copy1
```

```python
# Save client_Table as an csv file without the index column
client_Table_copy1.to_csv('client_Table.csv', index=False)
```

```python
# Save client_Table as an excel file for easy assess on excel with the index column
client_Table_copy1.to_excel('client_Table.xlsx', index=True)
```

# Step 3: SQL Code Implementation

```python
# Define SQL code for creating tables
table_creation_sql = """
-- SQL code for creating tables
-- Create the Client table
CREATE TABLE Client (
     client_id INT PRIMARY KEY,
     age INT,
     job VARCHAR(255),
     marital VARCHAR(255),
     education VARCHAR(255),
     credit_default BOOLEAN,
     housing BOOLEAN,
     loan BOOLEAN
);

-- Create the Campaign table
CREATE TABLE Campaign (
    campaign_id VARCHAR(10) PRIMARY KEY,
    client_id INT,
    numbers_contacts INT,
    contacts_duration INT,
    pdays INT,
    previous_campaign_contacts INT,
    previous_outcome BOOLEAN,
    campaign_outcome BOOLEAN,
    last_contact_date DATE,
    FOREIGN KEY (client_id) REFERENCES Client(client_id)
);

-- Create the Economics table
CREATE TABLE Economics (
    client_id INT,
    emp_var_rate DECIMAL(18, 2),
    cons_price_idx DECIMAL(18, 2),
    euribor_three_months DECIMAL(18, 2),
    number_employed DECIMAL(18, 2),
    FOREIGN KEY (client_id) REFERENCES Client(client_id)
);
"""

# Define SQL code for populating tables with data
tables_populating_sql = """
-- SQL code for populating tables with data
-- Populate the Client table with data from 'client_data.csv'
COPY Client FROM 'client_data.csv' DELIMITER ',' CSV HEADER;

-- Populate the Campaign table with data from 'campaign_data.csv'
COPY Campaign FROM 'campaign_data.csv' DELIMITER ',' CSV HEADER;

-- Populate the Economics table with data from 'economics_data.csv'
COPY Economics FROM 'economics_data.csv' DELIMITER ',' CSV HEADER;
"""

# Save SQL code as multiline string variables
sql_code = {
    'Table_Create_sql': table_creation_sql,
    'populate_tables_sql': tables_populating_sql
}

# Display the SQL code for the bank to use
sql_code['Table_Create_sql']
sql_code['populate_tables_sql']
```
